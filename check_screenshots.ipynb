```python
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

# Read your storage constants
client_id = os.getenv('CLIENT_ID')
storage_account_url = os.getenv('STORAGE_ACCOUNT_URL')
storage_container = os.getenv('STORAGE_CONTAINER')
```

```python
from azure.identity import ManagedIdentityCredential
from azure.storage.blob import BlobServiceClient

# Authenticate using managed identity
credentials = ManagedIdentityCredential(client_id=client_id)

# Create a BlobServiceClient
blob_service_client = BlobServiceClient(account_url=storage_account_url, credential=credentials)

# Get the container client
container_client = blob_service_client.get_container_client(storage_container)

# List all blobs (files) in the container
blob_list = container_client.list_blobs()
for blob in blob_list:
    print(blob.name)
```
